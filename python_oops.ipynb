#Q6. Explain the concept of inheritance in OOPH?

Ans. Inheritance in Object-Oriented Programming (OOP) allows a new class to inherit properties and behaviors from an existing class. This concept promotes code reusability and establishes an "is-a" relationship between classes.
The class from which properties and methods are inherited is called the base class, while the class that inherits from the base class is known as the derived class. Inheritance is important for code reusability, hierarchy establishment, enhanced maintainability, extensibility, and serving as a foundation for polymorphism.

Common types of inheritance include single inheritance, where a class inherits from one base class, and multiple inheritance, where a class inherits from multiple base classes. Other types are multilevel inheritance, where a class inherits from a derived class, and hierarchical inheritance, where multiple classes inherit from a single base class. Hybrid inheritance is a combination of two or more types.

Inheritance can have limitations, such as tight coupling between classes, the fragile base class problem, and increased complexity with deep hierarchies. The diamond problem can also arise in languages supporting multiple inheritance. Best practices for using inheritance include using it for clear "is-a" relationships, considering composition over inheritance, utilizing interfaces and abstract classes, and keeping hierarchies shallow. Inheritance is a fundamental concept for creating efficient, modular, and maintainable software.


#Q7. What is polymorphism in OOP+?

Ans. Polymorphism in object-oriented programming:

Polymorphism, a core principle of Object-Oriented Programming (OOP), means "many forms". It allows objects of different classes to be treated as objects of a common superclass or interface, exhibiting multiple forms or behaviors based on their class or interface. This promotes flexibility, code reusability, and extensibility.

Types of polymorphism:

Polymorphism has two main types:

Compile-time polymorphism (static binding or early binding): Resolved during compilation, the compiler determines which method or operator to use based on parameter types. This includes method and operator overloading.

Runtime polymorphism (dynamic binding or late binding): Resolved during execution, the system determines the method implementation based on the object's actual type. This is achieved through method overriding and virtual functions.

Polymorphism relies on inheritance and method overriding. Its benefits include code reusability, flexibility, extensibility, abstraction, and simplified code.

In Python, the constructor is defined using the __init__() method. It's not explicitly called but is automatically invoked when you create an object of the class.

Key aspects

Purpose: The main goal of the __init__() method is to set the initial state of an object by assigning values to its attributes.

self parameter: The __init__() method, and all instance methods in a class, must take self as their first parameter. This self refers to the instance of the class being created and allows you to access its attributes and methods.
Automatic invocation: You don't call the constructor explicitly; Python calls it automatically when you create an object of the class.

Types of constructors:

Non-parameterized (Default) constructor: This type of constructor does not accept any arguments apart from the mandatory self. It sets default values for object attributes during creation.

Parameterized constructor: This constructor takes arguments in addition to self, allowing you to initialize object attributes with custom values when the object is created.

Constructor overloading (simulation): Python doesn't directly support constructor overloading in the same way some other languages do (like Java or C++). However, you can achieve a similar effect using default arguments, variable-length arguments (*args, **kwargs), or class methods marked with the @classmethod decorator.

In essence, constructors in Python ensure that newly created objects are set up correctly with their necessary attributes, contributing to well-structured and maintainable code.

Method overriding in Object-Oriented Programming (OOP) allows a subclass to provide its own version of a method already present in its superclass. This is useful when the subclass needs different behavior than the superclass.

Dynamic Method Dispatch: The specific method executed is determined at runtime based on the object's actual type, not the reference variable's type. This is also known as runtime polymorphism or late binding.

Advantages

Customized functionality: Subclasses can implement inherited methods to suit their needs.

Code reusability: It allows for reusing method structures while altering behavior in subclasses.

Flexibility and extensibility: It supports building upon existing code without changing the superclass.

Example

Consider a Shape class with a calculate_area() method. Subclasses like Circle and Square can override this method to calculate their specific areas.
Method overriding is essential for creating flexible and dynamic OOP code, promoting reuse, polymorphism, and customization within an inheritance structure.


#Q13. What is a property decorator in Python+?

Ans. In Python, the @property decorator is a built-in feature used to manage class attributes with more control and flexibility. It allows you to define methods that behave like attributes, meaning you can access, modify, or delete them using the standard dot notation (object.attribute) instead of explicit method calls (object.get_attribute(), object.set_attribute(value)).

How it works

The @property decorator is essentially a shorthand way to define getter, setter, and deleter methods for an attribute, according to Real Python.
@property (getter): You define a method and decorate it with @property. This method acts as the getter for the attribute. Whenever you access the attribute, this method is automatically called, says Medium.

@<attribute_name>.setter (setter): To allow modifications to the attribute, you define another method with the same name as the getter, but decorated with @<attribute_name>.setter. This method is invoked automatically when you assign a new value to the attribute.

@<attribute_name>.deleter (deleter): To enable deletion of the attribute, you define a third method, also with the same name, and decorate it with @<attribute_name>.deleter. This method is called when you use the del keyword on the attribute.

Benefits of using the @property decorator
Encapsulation: It allows you to control how attributes are accessed and modified, enforcing rules and validation logic, according to dev.to.
Readability and Pythonic syntax: It allows you to interact with attributes using a simpler, more intuitive syntax, making your code easier to read and understand.

Data validation: You can include validation logic within the setter method to ensure that attribute values meet specific criteria, says Medium.

Backward compatibility: If you initially use direct attribute access but later need to add getter/setter logic, you can easily convert the attribute into a property without breaking existing code that interacts with the attribute.
Computed properties: You can define properties whose values are calculated dynamically based on other attributes.

class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):
        """Getter for the radius property."""
        return self._radius

    @radius.setter
    def radius(self, value):
        """Setter for the radius property with validation."""
        if value <= 0:
            raise ValueError("Radius must be positive.")
        self._radius = value

    @radius.deleter
    def radius(self):
        """Deleter for the radius property."""
        del self._radius

    @property
    def area(self):
        """Compute the area of the circle."""
        return 3.14 * self._radius ** 2

# Usage
circle = Circle(5)
print(circle.radius)  # Accessing the getter method like an attribute. Output: 5
print(circle.area)   # Accessing the computed property. Output: 78.5

circle.radius = 10  # Calling the setter method like an attribute assignment
print(circle.area)   # Output: 314.0

del circle.radius  # Invoking the deleter method
# print(circle.radius)  # This would raise an AttributeError because the attribute has been deleted.

#Q15. What is an abstract class in Python+?

Ans. In Python, an abstract class acts as a blueprint or a template for other classes to follow. It defines a common interface for a group of related classes, ensuring that they adhere to a consistent structure and behavior. Unlike regular classes, abstract classes cannot be instantiated directly. Instead, they are designed to be inherited by subclasses which then provide concrete implementations for the abstract methods defined in the abstract class.


Key features and characteristics

Uses the abc module: Python doesn't directly support abstract classes in the same way some other object-oriented languages do. To create abstract classes in Python, you use the abc (Abstract Base Classes) module, which provides the necessary tools like ABC (the abstract base class itself) and the @abstractmethod decorator.

Cannot be instantiated: You cannot create an object directly from an abstract class. If you try, Python will raise a TypeError. This restriction ensures that the abstract class acts purely as a blueprint and is never used in an incomplete state.

Contains abstract methods: Abstract classes can have one or more abstract methods. An abstract method is declared but does not have an implementation (its body is usually empty, using the pass keyword). The @abstractmethod decorator is used to mark a method as abstract.

Requires subclasses to implement abstract methods: Any class that inherits from an abstract class must provide a concrete implementation for all the abstract methods defined in the abstract parent class. If a subclass fails to implement an abstract method, attempting to instantiate that subclass will result in a TypeError.

Can have concrete methods: Abstract classes can also include concrete methods (methods with implementations). These methods can be inherited by subclasses, promoting code reuse.

Enforces a contract: Abstract classes enforce a contract, ensuring that subclasses provide specific functionalities. This promotes consistency across different parts of a program.

Supports polymorphism: Abstract classes facilitate polymorphism by allowing code to operate on objects of different subclasses uniformly, as long as those subclasses implement the defined interface.

In Object-Oriented Programming (OOP), particularly in Python, understanding the distinction between class variables and instance variables is crucial for effective class design and managing data within your programs.

Here's a breakdown of the key differences:

Class variables

Definition & Ownership: Defined within the class scope outside any methods, class variables are associated with the class itself and are owned by the class. There is a single copy shared among all instances.
Access & Modification: They can be accessed using the class name or an instance. Modifying a class variable through the class name affects all instances, while modifying through an instance creates a new instance variable for that specific instance, shadowing the class variable.
Use Cases: Ideal for data common to all instances, such as constants or default values.

Instance variables

Definition & Ownership: Typically defined within methods using the self keyword, instance variables are owned by each individual instance of the class. Each object has its own copy.

Access & Modification: Accessed only through an object. Changes to an instance variable in one object do not affect others.

Use Cases: Suitable for data specific and unique to each object


#Q18. What is multiple inheritance in Python+?

Ans. Multiple inheritance in Python

In Python, multiple inheritance is a feature that allows a class to inherit attributes and methods from more than one parent (or base) class. This means a single child class can combine and extend the functionalities of multiple parent classes, leading to increased code reuse and greater flexibility in designing complex class hierarchies.  

How it works

When defining a class with multiple inheritance, you list the base classes within the parentheses of the class definition, separated by commas.
A basic example is a class DerivedClass inheriting from BaseClass1 and BaseClass2. This allows an instance of DerivedClass to access methods from both parent classes. You can find a code example in the referenced documents.
Method resolution order (MRO)

The Method Resolution Order (MRO) is crucial in multiple inheritance as it dictates the search order for methods and attributes in the class hierarchy. This order resolves potential conflicts when multiple parent classes share the same method or attribute name. Python uses the C3 linearization algorithm to determine the MRO. The MRO can be viewed using the mro() method or the __mro__ attribute.

For instance, in a class D inheriting from B and C, where both B and C inherit from A, the MRO determines which version of a shared method is called. Python's MRO in this case would typically prioritize B over C if B is listed first in the inheritance. A code example illustrating this is available in the provided sources.

The diamond problem

The "diamond problem" occurs when a class inherits from two classes that share a common base class, leading to ambiguity if a method exists in both intermediate classes but is not overridden in the final class. Python's MRO effectively solves this problem by defining a consistent order for method resolution.

Use cases and considerations

Multiple inheritance offers advantages such as enhanced code reusability by allowing classes to inherit from multiple sources, reducing duplication. It also provides greater flexibility in class design by combining different functionalities. Mixins, which are classes providing specific methods to be incorporated into other classes, are often implemented using multiple inheritance to improve modularity.

However, multiple inheritance can increase code complexity, making it harder to understand and debug. Naming conflicts between methods in parent classes are also a potential issue. Careful design and understanding of the MRO are essential when using multiple inheritance. Composition, where a class uses instances of other classes instead of inheriting from them, can be an alternative approach with similar benefits.

In Python, __str__ and __repr__ are special "dunder" (double underscore) methods used to define how an object is represented as a string. While both return a string, they serve different purposes and are used in different contexts.  

__str__
Purpose: Provides a human-readable, informal representation of an object, intended for end-users.

When called: Automatically invoked by print(), str(), and format().

Output: Should be concise and easy for humans to understand, not necessarily a valid Python expression.

Purpose: Offers an information-rich, official representation for developers, useful for debugging and logging.

When called: Used by repr() and in the interactive Python interpreter.

Output: Ideally, a valid Python expression that could recreate the object.

Fallback: If __str__ is not defined, __repr__ is used when str() or print() are called.

Key differences summarised
Feature 	__str__	__repr__
Audience	End-users	Developers (for debugging and logging)
Readability	Human-readable and informal	Formal and unambiguous
Output	Concise and informative for users.	Detailed representation (can be used to recreate object).

Called by	print(), str(), format()	repr(), interactive interpreter
Fallback	__repr__ is used if __str__ is not defined.

Implement __repr__ for your custom classes to ensure a useful string representation for developers.

In Python, the super() function is a powerful built-in tool that plays a crucial role in managing inheritance within Object-Oriented Programming (OOP) by providing access to the methods and attributes of a parent or sibling class from a child or subclass. It essentially allows subclasses to extend and customize the functionality inherited from their parent classes, enabling more maintainable, flexible, and robust code.
Here's a breakdown of the significance of the super() function:

1. Facilitating method overriding and extension
super() allows a subclass to call a method from its parent class even if the subclass has overridden that method, according to GeeksforGeeks.
This means you can extend the functionality of the parent's method by calling it first using super() and then adding specific behavior in the subclass, rather than completely replacing it.
It promotes code reusability and ensures that the inherited behavior is properly integrated into the subclass.

2. Proper initialization in inheritance hierarchies
When a subclass overrides the __init__() (constructor) method of its parent class, it's essential to ensure that the parent's initialization is also executed, according to Sourcebae.
super().__init__() is used in the subclass's __init__() method to call the parent's constructor, initializing any attributes or behaviors defined in the parent class.

Here's a breakdown of the key distinctions:

1. Implicit first argument
@classmethod: Automatically receives the class itself as the first argument, conventionally named cls. This allows the method to access and modify class-level attributes.
@staticmethod: Does not receive any implicit first argument like self (instance) or cls (class). It functions like a regular function, only bound to the class for organizational purposes.

2. Access to class and instance state
@classmethod: Can access and modify class-level attributes and call other class methods, because it takes cls as an argument.
@staticmethod: Cannot access or modify class or instance attributes directly, since it has no knowledge of the class or instance state. It can only operate on parameters passed to it.

3. Primary use cases
@classmethod: Typically used for:
Factory methods: These methods return instances of the class, often providing alternative ways to create objects based on different inputs. For example, a Person class might have a class method from_birth_year to create Person objects from a birth year.
Operations that involve the class as a whole: Methods that interact with or modify class-level data.
@staticmethod: Typically used for:
Utility functions: Helper functions that are logically related to the class but don't need access to class or instance attributes. For example, a MathUtils class could have a static method calculate_factorial.
Methods that encapsulate functionality within the class's namespace without being tied to instances.

4. Inheritance behavior
@classmethod: Behaves polymorphically. When a subclass inherits a class method, the cls argument in the method will refer to the subclass itself, ensuring the correct type of object is created if it's a factory method.
@staticmethod: While static methods can be inherited and overridden, their behavior doesn't change based on the subclass. They don't have an implicit cls argument to reference the inheriting class.
In summary
Feature 	@classmethod	@staticmethod
Implicit First Argument	cls (class itself)	None

Common Use Cases	Factory methods, operations on class-level data.	Utility functions, grouping related functions.

#Q24. What is method chaining in Python OOP+?

Ans. Method chaining in Python OOP
Method chaining in Python object-oriented programming (OOP) is a technique that enables calling multiple methods on the same object in a single line of code, according to GeeksforGeeks. This approach makes code more concise, readable, and often easier to maintain by eliminating the need for intermediate variables.

How method chaining works

For method chaining to function, each method in the chain must return an object, typically the same object (using self in Python) that the methods are being called upon.

Here's a simple example:


```python
class Calculator:
    def __init__(self, value=0):
        self.value = value

    def add(self, num):
        self.value += num
        return self  # Return self to enable chaining

    def multiply(self, num):
        self.value *= num
        return self  # Return self to enable chaining

    def subtract(self, num):
        self.value -= num
        return self  # Return self to enable chaining

# Using method chaining
result = Calculator().add(5).multiply(3).subtract(2)
print(result.value)  # Output: 13
```

In this example, the add(), multiply(), and subtract() methods all return self, the instance of the Calculator object itself. This allows us to chain the method calls, resulting in the concise expression Calculator().add(5).multiply(3).subtract(2).

Benefits of method chaining

Conciseness and Readability: Chained methods can make code shorter and easier to read, as the operations flow in a natural, sequential manner, often resembling a natural language sentence.

Reduced Boilerplate: It eliminates the need to create and manage intermediate variables for storing the results of each operation, making the code cleaner and less verbose.

Fluent Interface Design: Method chaining can be used to create APIs (Application Programming Interfaces) that are intuitive and expressive, making them easier to use and understand for developers, says Analytics Vidhya.
Expressiveness: Complex operations can be expressed in a single line, clearly demonstrating the intention of the code.

#Q25. What is the purpose of the __call__ method in Python?

Ans. The purpose of __call__ method in Python

The __call__ method in Python, also known as a "dunder" or "magic" method, allows you to make instances of a class callable like functions. When you define this method within a class, you can call an object of that class using the standard function call syntax (e.g., obj(arg1, arg2)), and it will automatically execute the code within the __call__ method.

Key functionalities and use cases

1. Creating callable instances
The primary purpose is to enable class instances to behave like functions, providing a flexible way to add functionality to your objects.
This allows you to encapsulate behavior within objects and then "call" that behavior directly on the instance, notes Real Python.

2. State retention and function factories
__call__ allows objects to retain state between calls. This means the object can remember information from previous calls and use it in subsequent calls.
This enables the creation of "function factories" where instances act like functions with customizable behavior based on the initial arguments passed to the class constructor (__init__).

3. Implementing decorators
__call__ is essential for implementing class-based decorators. A class acting as a decorator needs its instances to be callable to wrap and modify the behavior of functions or methods.

4. Designing convenient APIs (Application Programming Interfaces)
By implementing __call__, you can create intuitive and easy-to-use APIs. For instance, an object can be configured during instantiation and then called directly to perform its main operation.

5. Strategy pattern implementation
__call__ is also valuable when implementing the strategy design pattern, allowing you to define different algorithms as separate classes, each with its __call__ method, and dynamically select which algorithm to use at runtime.
__call__ vs __init__

It's important to differentiate __call__ from __init__:

__init__ is called when you create an instance of the class (e.g., my_object = MyClass(arg)), initializing the object's attributes.
__call__ is invoked when you call an existing instance of the class as a function (e.g., my_object(arg)), executing the code defined within the __call__ method.

python

class Multiplier:
    def __init__(self, factor):
        self.factor = factor  # Store the factor as instance state

    def __call__(self, x):
        return x * self.factor # Multiplies the input by the stored factor

# Create callable instances with different factors
double = Multiplier(2)
triple = Multiplier(3)

print(double(5))  # Output: 10
print(triple(5))  # Output: 15

Use code with caution.

In this example, the Multiplier class creates callable instances that multiply a number by a factor defined during instantiation.

```python
#Q7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.

#Ans.

class MathOperations:
    # A class method to add two numbers.
    # Class methods take 'cls' as their first argument, which refers to the class itself.
    # They are typically used for operations that involve the class or its state.
    @classmethod
    def add_numbers(cls, num1, num2):
        """
        Adds two numbers using a class method.
        """
        print(f"Adding numbers using a class method (via {cls.__name__}):")
        return num1 + num2

    # A static method to subtract two numbers.
    # Static methods don't take 'self' (instance) or 'cls' (class) as their first argument.
    # They behave like regular functions but are logically grouped within the class.
    @staticmethod
    def subtract_numbers(num1, num2):
        """
        Subtracts two numbers using a static method.
        """
        print("Subtracting numbers using a static method:")
        return num1 - num2

# Call the class method using the class name
sum_result = MathOperations.add_numbers(10, 5)
print(f"Sum: {sum_result}")

# Call the static method using the class name
difference_result = MathOperations.subtract_numbers(10, 5)
print(f"Difference: {difference_result}")

# You can also call the methods using an instance, but it's more common to call them directly on the class for this type of operation
math_instance = MathOperations()
sum_result_instance = math_instance.add_numbers(20, 8)
print(f"Sum (via instance): {sum_result_instance}")
difference_result_instance = math_instance.subtract_numbers(20, 8)
print(f"Difference (via instance): {difference_result_instance}")
```

```python
#Q12.Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.

#Ans. class Student:
    def __init__ (self, name, grades):
        self.name = name  # Student's name (string)
        self.grades = grades  # List of grades (e.g., [85, 90, 78, 92])

    def average_grade(self):
        """
        Computes and returns the average of the student's grades.
        Returns 0 if the grades list is empty to avoid division by zero errors.
        """
        if not self.grades:  # Check if the list of grades is empty
            return 0  # Return 0 or handle as appropriate for an empty list

        # Calculate the sum of grades and divide by the number of grades
        return sum(self.grades) / len(self.grades)

# Create a Student object
student1 = Student("Alice", [85, 90, 78, 92, 88])

# Compute and print the average grade
print(f"Student: {student1.name}")
print(f"Grades: {student1.grades}")
print(f"Average Grade: {student1.average_grade():.2f}") # Format to 2 decimal places

# Create another student with different grades
student2 = Student("Bob", [70, 75, 80])
print(f"\nStudent: {student2.name}")
print(f"Grades: {student2.grades}")
print(f"Average Grade: {student2.average_grade():.2f}")

# Demonstrate with an empty list of grades
student3 = Student("Charlie", [])
print(f"\nStudent: {student3.name}")
print(f"Grades: {student3.grades}")
print(f"Average Grade: {student3.average_grade():.2f}")
```

```python
#Q13.Create a class Rectangle with methods set_dimensions() to set the dimensions and area() to calculate the area.

#ans.

class Rectangle:
    def __init__(self, length=0, width=0):
        # Initialize length and width with default values
        # or with the values provided when creating an object
        self.length = length
        self.width = width

    def set_dimensions(self, length, width):
        """
        Sets the length and width of the rectangle.
        Includes basic validation to ensure dimensions are positive.
        """
        if length > 0 and width > 0:
            self.length = length
            self.width = width
            print(f"Rectangle dimensions set to: Length = {self.length}, Width = {self.width}")
        else:
            print("Length and width must be positive values. Dimensions not updated.")

    def area(self):
        """
        Calculates and returns the area of the rectangle.
        The area of a rectangle is calculated using the formula, area of rectangle = length × width.
        """
        if self.length > 0 and self.width > 0:
            return self.length * self.width
        else:
            return 0  # Return 0 if dimensions are not set or invalid

# Create a Rectangle object
rect1 = Rectangle()
print(f"Initial Area of rect1: {rect1.area()}") # Output: Initial Area of rect1: 0

# Set dimensions using set_dimensions()
rect1.set_dimensions(5, 10)
print(f"Area of rect1: {rect1.area()}")  # Output: Area of rect1: 50

# Create another Rectangle object with initial dimensions
rect2 = Rectangle(7, 4)
print(f"\nArea of rect2: {rect2.area()}") # Output: Area of rect2: 28

# Attempt to set invalid dimensions
rect1.set_dimensions(-2, 8) # Output: Length and width must be positive values. Dimensions not updated.
print(f"Area of rect1 after invalid attempt: {rect1.area()}") # Output: Area of rect1 after invalid attempt: 50
```

```python
#Q14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.

#ans. class Employee:
    def__init__(self, name, hourly_rate):
        self.name = name
        self.hourly_rate = hourly_rate

    def calculate_salary(self, hours_worked):
        """
        Calculates the employee's salary based on hours worked and hourly rate.
        Assumes no overtime for simplicity in this example.
        """
        if hours_worked < 0:
            print("Hours worked cannot be negative.")
            return 0
        return hours_worked * self.hourly_rate

# Derived class Manager, inheriting from Employee
class Manager(Employee):
    def __init__(self, name, hourly_rate, bonus_percentage):
        # Call the parent class (Employee) constructor using super()
        super().__init__(name, hourly_rate)
        if 0 <= bonus_percentage <= 1:
            self.bonus_percentage = bonus_percentage  # Bonus as a percentage (e.g., 0.1 for 10%)
        else:
            print("Bonus percentage must be between 0 and 1. Setting to 0.")
            self.bonus_percentage = 0

    def calculate_salary(self, hours_worked):
        """
        Overrides the calculate_salary method to include a bonus for managers.
        """
        # Calculate the base salary using the parent's calculate_salary method
        base_salary = super().calculate_salary(hours_worked)
        bonus_amount = base_salary * self.bonus_percentage
        total_salary = base_salary + bonus_amount
        return total_salary

# Create an Employee object
employee1 = Employee("Alice", 25)
hours_worked_employee = 160
employee_salary = employee1.calculate_salary(hours_worked_employee)
print(f"{employee1.name}'s salary: ₹{employee_salary:.2f}")

# Create a Manager object
manager1 = Manager("Bob", 35, 0.1) # 10% bonus
hours_worked_manager = 160
manager_salary = manager1.calculate_salary(hours_worked_manager)
print(f"{manager1.name}'s salary: ₹{manager_salary:.2f}")

# Create another Manager with a different bonus
manager2 = Manager("Charlie", 40, 0.15) # 15% bonus
manager_salary2 = manager2.calculate_salary(160)
print(f"{manager2.name}'s salary: ₹{manager_salary2:.2f}")

# Demonstrate the case where no bonus is provided
manager3 = Manager("David", 30, 0)
manager_salary3 = manager3.calculate_salary(160)
print(f"{manager3.name}'s salary: ₹{manager_salary3:.2f}")

# Demonstrate invalid hours worked
print(f"\nAttempting to calculate salary for negative hours worked for {employee1.name}:")
invalid_salary = employee1.calculate_salary(-10)
print(f"Result: ₹{invalid_salary:.2f}")
```

```python
#Q15. Create a class Product with attributes name, price, and quantity. Implement a method total_price() that calculates the total price of the product.

#ans

class Product:
    def __init__(self, name, price, quantity):
        self.name = name          # Name of the product (string)
        self.price = price        # Price per unit (float or int)
        self.quantity = quantity  # Number of units (int)

    def total_price(self):
        """
        Calculates and returns the total price of the product
        based on its price per unit and quantity.
        """
        # Ensure price and quantity are valid before calculation
        if self.price >= 0 and self.quantity >= 0:
            return self.price * self.quantity
        else:
            print("Invalid price or quantity. Cannot calculate total price.")
            return 0  # Return 0 or handle error appropriately

# Create Product objects
product1 = Product("Laptop", 1200.50, 2)
product2 = Product("Mouse", 25.00, 5)
product3 = Product("Keyboard", 75.99, 1)

# Calculate and print the total price for each product
print(f"Product: {product1.name}")
print(f"Price per unit: ₹{product1.price:.2f}")
print(f"Quantity: {product1.quantity}")
print(f"Total Price: ₹{product1.total_price():.2f}\n")

print(f"Product: {product2.name}")
print(f"Price per unit: ₹{product2.price:.2f}")
print(f"Quantity: {product2.quantity}")
print(f"Total Price: ₹{product2.total_price():.2f}\n")

print(f"Product: {product3.name}")
print(f"Price per unit: ₹{product3.price:.2f}")
print(f"Quantity: {product3.quantity}")
print(f"Total Price: ₹{product3.total_price():.2f}\n")

# Demonstrate with invalid data
invalid_product = Product("Broken Item", -10.0, 3)
print(f"Product: {invalid_product.name}")
print(f"Price per unit: ₹{invalid_product.price:.2f}")
print(f"Quantity: {invalid_product.quantity}")
print(f"Total Price: ₹{invalid_product.total_price():.2f}")
```

```python
#Q16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.

#ans.

from abc import ABC, abstractmethod

# Abstract Base Class
class Animal(ABC):  # Inherits from ABC to be an abstract class
    @abstractmethod  # Decorator to define an abstract method
    def sound(self):
        """
        Abstract method to be implemented by derived classes.
        Each animal will have its own unique sound.
        """
        pass  # Abstract methods do not have an implementation in the abstract class

# Derived Class: Cow
class Cow(Animal):
    def sound(self):
        """
        Implements the sound() method for a Cow.
        """
        print("Moo!")

# Derived Class: Sheep
class Sheep(Animal):
    def sound(self):
        """
        Implements the sound() method for a Sheep.
        """
        print("Baa!")

# --- Demonstration ---
# You cannot directly instantiate an abstract class
try:
    generic_animal = Animal()
except TypeError as e:
    print(f"Error: {e}") # Output: Error: Can't instantiate abstract class Animal with abstract method sound

# Create instances of the derived classes
cow = Cow()
sheep = Sheep()

# Call the sound() method on each instance
cow.sound()   # Output: Moo!
sheep.sound() # Output: Baa!

# Polymorphism in action: A list of Animal objects
animals = [Cow(), Sheep()]

print("\n--- Iterating through animals ---")
for animal in animals:
    animal.sound()
# Output:
# Moo!
# Baa!
```

```python
#Q17. Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that returns a formatted string with the book's details.

#Ans.

class Book:
    def __init__(self, title, author, year_published):
        self.title = title                # Title of the book (string)
        self.author = author              # Author of the book (string)
        self.year_published = year_published # Year the book was published (integer)

    def get_book_info(self):
        """
        Returns a formatted string containing the book's title, author, and year published.
        """
        return f"Title: {self.title}, Author: {self.author}, Year Published: {self.year_published}"

# Create instances of the Book class
book1 = Book("The Hitchhiker's Guide to the Galaxy", "Douglas Adams", 1979)
book2 = Book("Pride and Prejudice", "Jane Austen", 1813)
book3 = Book("1984", "George Orwell", 1949)

# Get and print the book information using the get_book_info() method
print(book1.get_book_info())
print(book2.get_book_info())
print(book3.get_book_info())
```

```python
#Q18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.

#Ans.


class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price
        print(f"A house at {self.address} with a price of ${self.price:,} is created.") # Informative message, formatted with commas for price

    def get_house_info(self):
        return f"Address: {self.address}, Price: ${self.price:,}"

class Mansion(House): # Mansion inherits from House
    def __init__(self, address, price, number_of_rooms):
        # Call the constructor of the parent class (House)
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms # Add the new attribute specific to Mansion
        print(f"This mansion has {self.number_of_rooms} rooms.") # Informative message

    def get_mansion_info(self):
        # Access attributes from both parent and child classes
        return f"{super().get_house_info()}, Number of Rooms: {self.number_of_rooms}"

# Create instances
my_house = House("123 Main St", 500000)
my_mansion = Mansion("456 Grand Ave", 2500000, 15)

# Demonstrate accessing attributes and methods
print("\n--- House Information ---")
print(my_house.get_house_info())

print("\n--- Mansion Information ---")
print(my_mansion.get_house_info()) # Mansion inherits get_house_info() from House.
print(my_mansion.get_mansion_info())
```
